```python
Beta = 0.5  #(fixed)
clips = [0.2, 0.3, 0.4]
l_noises = [0.15, 0.2, 0.25, 0.3, 0.35]
for clip in clips:
    var_name = 'epsilon_{}_ML_1M'.format(clip)
    locals()[var_name] = []
    for l_noise in l_noises:
        locals()[var_name].append(2*(1 + Beta)*clip/l_noise)
    np.save('./{}.npy'.format(var_name), locals()[var_name])
```
